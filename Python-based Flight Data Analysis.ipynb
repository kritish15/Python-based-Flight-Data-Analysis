```python
import pandas as pd
```

```python
# Load the CSV file into a DataFrame
data = pd.read_csv('flight_data.csv')

# Display the first few rows of the DataFrame
print(data.head())
```

```python
data['DepartureTime'] = pd.to_datetime(data['DepartureTime'])
data['ArrivalTime'] = pd.to_datetime(data['ArrivalTime'])
data['FlightDuration'] = (data['ArrivalTime'] - data['DepartureTime']).dt.total_seconds() / 60
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up Seaborn style
sns.set(style="whitegrid")

# Create a histogram of flight durations
plt.figure(figsize=(10, 6))
sns.histplot(data['FlightDuration'], bins=20, kde=True)
plt.xlabel('Flight Duration (minutes)')
plt.ylabel('Frequency')
plt.title('Distribution of Flight Durations')
plt.show()
```
